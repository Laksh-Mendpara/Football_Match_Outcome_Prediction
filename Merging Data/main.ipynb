```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.gridspec as gridspec
from numpy import random
import scipy.stats as scipy

from IPython.display import display
```

```python
data1 = pd.read_csv('total_dataset_v02.csv')
data1.tail()
```

```python
## if date is in format of DD/MM/YY then converting it to DD/MM/YYYY

def date_format(date):
    if date == '':
        return ''
    else:
        date = date.split('/')
        if len(date[2]) == 2:
            date[2] = '20' + date[2]
        return date[0] + '/' + date[1] + '/' + date[2]
```

```python
data1['Date'] = data1['Date'].apply(date_format)
data1.tail()
```

```python
## converting from DD/MM/YYYY to YYYY-MM-DD

def date_format_std(date):
    if date == '':
        return ''
    else:
        date = date.split('/')
        return date[2] + '-' + date[1] + '-' + date[0] + ' 00:00:00'
```

```python
data1['Date'] = data1['Date'].apply(date_format_std)
data1.tail()
```

```python
data2 = pd.read_csv('temp_data.csv')
data2.tail()
```

```python
# ## if date and (home_team_name of data1 & HomeTeam of data2) and (away_team_name of data1 & AwayTeam of data2) of both the datasets are same then merging them at their row and storing it in a new dataset

data3 = pd.merge(data1, data2, how='inner', left_on=['Date', 'HomeTeam', 'AwayTeam'], right_on=['date', 'home_team_long_name', 'away_team_long_name'])
```
